## tipos de productividad:

***Productividad parcial: ***

es la razon entre la cantidad producida y un solo tipo de insumo.

***Productividad multifactorial:***

es la productividad encontrada cono la suma de los factores, insumos

***Produsctividad total:***

Se entiende como la produccion neta con todos los factores de insumos.

```python
# Importamos las librerías necesarias
import pandas as pd

# Pedimos al usuario que ingrese los datos de entrada
input_data = {}
input_data['tiempo'] = float(input('Ingresa el tiempo total empleado (en horas): '))
input_data['cantidad'] = float(input('Ingresa la cantidad de productos o servicios producidos: '))
input_data['trabajadores'] = int(input('Ingresa el número total de trabajadores empleados: '))

# Calculamos la productividad laboral
output_data = {}
output_data['produccion_por_trabajador'] = input_data['cantidad'] / input_data['trabajadores']
output_data['produccion_por_hora'] = input_data['cantidad'] / input_data['tiempo']
output_data['productividad'] = output_data['produccion_por_trabajador'] / output_data['produccion_por_hora']

# Creamos un DataFrame para mostrar los resultados
df = pd.DataFrame(output_data, index=['Resultado'])

# Mostramos los resultados
print(df)
```

#Unidad 2: Organizacion de la produccion

- Sistema de organizacion "5s"

-Esta estrategia aplica a empresas de diferentes actividades económicas y tamaños, en áreas de oficinas, almacenes, talleres, plantas de producción, entre otras. En términos de su implementación es flexible, por lo que puede ser aplicada en una sola área de la empresa, a modo de “proyecto piloto”, o en todas las áreas, de manera simultánea.


La correcta aplicación de las 5S contribuirá a generar impactos importantes en:

 ### Aumento de productividad

1. Reducir Lead-times en procesos de manufactura y servicios, mejorando
 tiempos de entrega.
2. Reducir tiempos muertos en fábricas, talleres y oficinas, mediante aplicación de
 conceptos de gerencia visual o localización por defecto.
3. Reducir tiempos de alistamientos (set-up) en máquinas y equipos.
4. Reducir tiempos de acceso a materiales, documentos, herramientas y otros
 recursos utilizados en los procesos.

### Aumento de la calidad

1. Reducir errores humanos en procesos, por consiguiente, el porcentaje de unidades
 defectuosas en productos y servicios.
2. Reforzar estandarización de los procesos de manufactura y/o servicios.


### Reducción de costos

1. Mejorar gestión de inventarios.
2. Disminuir pérdidas por obsolescencia de inventario.
3. Disminuir riesgos potenciales de accidentes laborales, en consecuencia,
 reducir costos asociados a los mismos, asi como, a enfermedades ocupacionales.
4. Eficientizar uso del espacio físico en las instalaciones (Economía de espacio).
5. Extender vida útil de herramientas y equipos.

### Elevar moral del trabajador

1. Fomentar empoderamiento del personal involucrado, a fin de promover niveles
 de sinergia que faciliten el proceso de implementación.
2. Dignificar el puesto de trabajo.
3. Fomentar identificación y compromiso del personal, con equipos, herramientas,
 instalaciones y recursos de la empresa en general.
4. Fomentar cooperación y trabajo en equipo.


Se denominan como las “5S”, por estar basadas en la aplicación de cinco (5) conceptos
o principios de acción, cuyos términos originales en el
idioma japonés comienzan con la letra S.
- Estos principios, una vez aplicados al ambiente de trabajo, generan transformaciones físicas
que impactan positivamente la productividad de las operaciones que se ejecutan el mismo.

### DEFINICIONES DE LAS 5S

1. SEIRI = SELECCIONAR O CLASIFICAR
Separar elementos innecesarios de los que son necesarios. Descarte lo innecesario.
2. SEITON = ORGANIZAR
Colocar lo necesario en lugares fácilmente accesibles, según la frecuencia y secuencia
de uso. ¡Un lugar para cada cosa y cada cosa en su lugar!
3. SEISO = LIMPIAR
Limpiar completamente el lugar de trabajo, de tal manera que no haya polvo, ni grasa
en máquinas, herramientas, pisos, equipos, etc.
4. SEIKETSU = ESTANDARIZAR O MANTENER
Estandarizar la aplicación de las (3 S) anteriores, de tal manera que la aplicación de
éstas se convierta en una rutina o acto reflejo.
5. SHITSUKE = DISCIPLINAR
Entrenar a la gente para que aplique con disciplina las buenas prácticas de orden y
limpieza.

#### Definicion:
son representaciones gráficas de las etapas de un proceso. Se usan para visualizar las actividades, su secuencia y las relaciones entre ellas.

Existen diferentes tipos de diagramas de procesos, como:

1. Diagrama de flujo: Representa el flujo de trabajo con símbolos estandarizados (rectángulos, diamantes, flechas). Se usa para documentar procesos y detectar ineficiencias.

2. Diagrama SIPOC: Muestra los proveedores, entradas, proceso, salidas y clientes de un proceso. Se usa para tener una visión general del proceso y sus límites.

3. Diagrama de carriles: Similar al diagrama de flujo, pero organiza las actividades por roles o departamentos. Se usa para visualizar la responsabilidad de cada parte en el proceso.


#### Pasos para crear un diagrama de proceso

1. Define el proceso: Determina el alcance del proceso que quieres representar.
2. Identifica las actividades: Lista todas las actividades involucradas en el proceso.
3. Determina la secuencia: Ordena las actividades en la secuencia correcta.
4. Dibuja el diagrama: Utiliza símbolos y flechas para representar las actividades y su flujo.
5. Añade información: Incluye información adicional como tiempos, responsables, etc.


#### SIPOC:

1. SIPOC es un acrónimo que significa:

- Suppliers (Proveedores): ¿Quiénes proveen los insumos al proceso?
- Inputs (Entradas): ¿Qué materiales, información o recursos se requieren para el proceso?
- Process (Proceso): ¿Cuáles son los pasos o actividades que transforman las entradas en salidas?
- Outputs (Salidas): ¿Cuáles son los productos, servicios o resultados del proceso?
- Customers (Clientes): ¿Quiénes reciben las salidas del proceso?



2. ¿Cuándo usar un diagrama SIPOC?

Al inicio de un proyecto de mejora de procesos para definir el alcance.
Para comprender un proceso complejo de forma rápida.
Para identificar áreas de mejora potencial.
Para comunicar el proceso a los stakeholders de manera clara.


3. ¿Cómo crear un diagrama SIPOC?

1. Define el proceso: Especifica el proceso que quieres analizar.
2. Identifica las salidas (Outputs): ¿Qué se produce al final del proceso?
3. Identifica los clientes (Customers): ¿Quiénes reciben las salidas?
4. Identifica las entradas (Inputs): ¿Qué se necesita para el proceso?
5. Identifica los proveedores (Suppliers): ¿Quiénes proveen las entradas?
6. Describe el proceso (Process): Lista los elementos del sistema y su interaccion

##### Ejercicio SIPOC uno:

La cafeteria:

La cafetería de la universidad ha recibido quejas sobre los largos tiempos de espera para recibir sus pedidos, especialmente durante las horas pico. Los estudiantes se quejan de que pierden mucho tiempo en la fila y que a veces la comida no está a la temperatura adecuada.

Como ingeniero industrial, se te ha encomendado la tarea de analizar el proceso de servicio de la cafetería y proponer mejoras.

**Contexto:**

La cafetería de la universidad es un espacio concurrido donde los estudiantes, profesores y personal administrativo acuden a comprar alimentos y bebidas. Ofrece una variedad de opciones, desde snacks y bebidas frías hasta almuerzos completos. La cafetería cuenta con un área de autoservicio con diferentes estaciones (bebidas, ensaladas, platos calientes, postres, etc.) y una fila para pagar.

Durante las horas pico (entre las 12 pm y 2 pm), la cafetería se llena de estudiantes que buscan almorzar entre clases. Esto genera largas filas y tiempos de espera, lo cual causa molestia entre los estudiantes. Además, algunos estudiantes se quejan de que la comida no está a la temperatura adecuada cuando llegan a sus mesas.

Las autoridades de la universidad están preocupadas por esta situación, ya que afecta la satisfacción de los estudiantes y la imagen de la cafetería. Por esta razón, han contratado a un equipo de ingenieros industriales para analizar el proceso de servicio y proponer mejoras.

La cafetería tiene un sistema de punto de venta computarizado.
El personal de la cafetería incluye cocineros, cajeros y personal de limpieza.
La cafetería se abastece de alimentos de proveedores externos y también tiene una cocina interna para preparar algunos alimentos.

##### Registro de procesos:

1. Observación directa: El analista observa el proceso en tiempo real, registrando las actividades, los tiempos y los recursos utilizados. Se pueden usar herramientas como diagramas de flujo, hojas de verificación o videos.
2. Entrevistas: Se conversa con las personas involucradas en el proceso para obtener información sobre las actividades, los problemas y las posibles soluciones.
3. Recopilación de datos: Se recopilan datos sobre el proceso, como tiempos de ciclo, tasas de defectos, costos, etc. Se pueden usar herramientas como cronómetros, formularios o sistemas de información.


##### Análisis de procesos:

Una vez que se ha registrado la información del proceso, se procede a analizarla para identificar:

1. Pasos innecesarios: Actividades que no agregan valor al proceso y pueden ser eliminadas.
2. Cuellos de botella: Puntos del proceso donde se acumulan las tareas y se generan retrasos.
3. Duplicidades: Actividades que se realizan más de una vez sin necesidad.
Desperdicios: Uso ineficiente de recursos, como materiales, tiempo o energía.
4. Riesgos: Posibles problemas que pueden afectar el proceso.


##### Herramientas para el análisis de procesos:

- Diagramas de flujo: Permiten visualizar el flujo de trabajo y las relaciones entre las actividades.
- Diagramas causa-efecto: Ayudan a identificar las causas de un problema.
Histogramas: Muestran la distribución de los datos y permiten detectar patrones.
- Gráficos de Pareto: Identifican los problemas más importantes que deben abordarse primero.

###### Beneficios del registro y análisis de procesos:

1. Mejora la eficiencia: Se eliminan actividades innecesarias y se optimizan los recursos.
2. Reduce los costos: Se minimizan los desperdicios y se optimizan los tiempos.
3. Aumenta la calidad: Se detectan y corrigen los errores, mejorando la calidad del producto o servicio.
4. Facilita la toma de decisiones: Se cuenta con información precisa para tomar decisiones informadas.
