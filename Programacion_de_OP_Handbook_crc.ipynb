**La programación de la producción** implica la asignación de recursos y la determinación de tiempos de producción para cada tarea necesaria para completar un pedido o producción. Esta programación puede realizarse a largo, mediano y corto plazo.

- **Planificación a largo plazo:** En esta fase, se realiza la previsión de la demanda, se planifica la capacidad de producción y se planifican los recursos necesarios para producir los bienes o servicios. Esta etapa generalmente cubre un período de uno a cinco años y es esencial para la planificación estratégica de la empresa.

- **Planificación a medio plazo:** En esta fase, se realiza la programación maestra de la producción, que implica la planificación de la cantidad y el momento de producción de cada producto o servicio. Esta etapa generalmente cubre un período de tres a seis meses y es esencial para la planificación táctica de la empresa.

- **Planificación a corto plazo:** En esta fase, se realiza la programación detallada de la producción, que implica la planificación diaria o semanal de las operaciones de producción. Esta etapa es esencial para la gestión operativa de la empresa y permite ajustar la producción en función de la demanda y los recursos disponibles.

## Planeacion agregada:
***

Según el libro *Dirección de la Producción y de Operaciones: Decisiones tácticas de Jay Heizer y Barry Render*, la planificación agregada es un proceso de planificación que se centra en la determinación de la cantidad y el momento de la producción para un período de tiempo específico. La planificación agregada tiene como objetivo satisfacer la demanda del mercado al mismo tiempo que se minimizan los costos de producción.

El proceso de planificación agregada implica la determinación de:
- la capacidad de producción necesaria para satisfacer la demanda del mercado
- la determinación de la cantidad de recursos necesarios para producir los bienes y servicios requeridos
- la elaboración de un plan de producción a largo plazo que equilibre la capacidad y la demanda.

La planificación agregada se utiliza en empresas que producen bienes y servicios que no son fácilmente almacenables o que tienen una vida útil limitada. Algunos ejemplos de empresas que utilizan la planificación agregada incluyen empresas manufactureras, empresas de servicios de transporte y empresas de servicios de atención médica.

### tecnicas para la planeacion
***

Existen varias técnicas que se pueden utilizar en la planificación agregada, entre las que se incluyen:

1. **Planificación de inventarios:** esta técnica implica mantener un inventario adecuado de productos para satisfacer la demanda prevista durante un período determinado. Se puede utilizar un modelo de inventario para determinar el nivel óptimo de inventario.

2. **Planificación de la capacidad:** esta técnica implica ajustar la capacidad de producción para satisfacer la demanda prevista. Se puede utilizar un modelo de programación lineal para determinar el nivel óptimo de capacidad.

3. **Planificación de la fuerza laboral:** esta técnica implica ajustar la fuerza laboral para satisfacer la demanda prevista. Se puede utilizar un modelo de programación lineal para determinar el nivel óptimo de fuerza laboral.

4. **Subcontratación:** esta técnica implica subcontratar la producción a otra empresa para satisfacer la demanda prevista. Se puede utilizar un modelo de programación lineal para determinar el nivel óptimo de subcontratación.

## TEORÍA DE LAS RESTRICCIONES
El rendimiento total, un concepto importante en las operaciones, es el número de unidades procesadas en la planta y vendidas. El rendimiento total es una diferencia crítica entre una empresa de éxito y una empresa fracasada. Esto ha llevado a un enfoque en las restricciones, algo que se ha popularizado gracias al libro:
- *The Goal*: A Process of Ongoing Improvement (La Meta: Un Proceso de Mejora Continua) de Eliyahu Goldratt y Jeff Cox8


Las restricciones pueden ser físicas (como la disponibilidad de proceso o de personal, las materias primas o los suministros) o inmateriales (como procedimientos, moral y formación). Reconocer y gestionar estas limitaciones a través de un proceso de cinco pasos es la base de la teoría de las restricciones:
1. Identificar las restricciones.
2. Elaborar un plan para superar las restricciones identificadas.
3. Concentrar recursos en lograr el paso 2.
4. Reducir los efectos de las restricciones descargando trabajo o aumentando capacidad. Asegurarse de que las restricciones son reconocidas por todos aquellos que pueden tener influencia sobre ellas.
5. Una vez que se haya superado un conjunto de restricciones, volver al primer paso e identificar nuevas restricciones.

````python
import simpy
import random

# Definir el número de máquinas
NUM_MAQUINAS = 6

# Tiempos de procesamiento (en minutos)
TIEMPOS_PROCESAMIENTO = {
    1: lambda: 10,  # Make all values callable
    2: lambda: 9,
    3: lambda: 8,
    4: lambda: random.uniform(5, 7),  # A(5,7)
    5: lambda: random.uniform(5, 7),  # A(5,7)
    6: lambda: random.uniform(5, 7)   # A(5,7)
}

# Tiempo total de simulación (en minutos)
TIEMPO_SIMULACION = 480  # 8 horas

# Lista para guardar los tiempos de espera de cada producto
tiempos_espera = []

class Producto:
    def __init__(self, nombre, env):
        self.nombre = nombre
        self.env = env
        self.tiempo_creacion = self.env.now

    def procesar(self, maquina, tiempo_procesamiento):
        # Solicitar la máquina
        with maquina.request() as req:
            yield req
            # Procesar el producto
            yield self.env.timeout(tiempo_procesamiento()) # Call the function to get the processing time
            print(f"{self.nombre} procesado en la máquina {maquina.count} en el tiempo {self.env.now}")

def proceso_produccion(env, nombre, maquinas):
    producto = Producto(nombre, env)
    for i in range(1, NUM_MAQUINAS + 1):
        tiempo_procesamiento = TIEMPOS_PROCESAMIENTO[i]
        yield env.process(producto.procesar(maquinas[i-1], tiempo_procesamiento))
    tiempos_espera.append(env.now - producto.tiempo_creacion)

def ejecutar_simulacion():
    env = simpy.Environment()
    maquinas = [simpy.Resource(env, capacity=1) for _ in range(NUM_MAQUINAS)]
    for i in range(10):  # Simular 10 productos
        env.process(proceso_produccion(env, f"Producto {i+1}", maquinas))
    env.run(until=TIEMPO_SIMULACION)

    # Calcular y mostrar estadísticas
    tiempo_espera_promedio = sum(tiempos_espera) / len(tiempos_espera)
    print(f"\nTiempo de espera promedio: {tiempo_espera_promedio:.2f} minutos")

if __name__ == "__main__":
    ejecutar_simulacion()
````

#### estrategia del DRUM BUFFER ROPE (DBR)

Definicion:

El método de cuerda amortiguadora de tambor (DBR) es una técnica de planificación y gestión de la producción que se basa en la teoría de las limitaciones. El objetivo del DBR es sincronizar el flujo de materiales y recursos con el ritmo del recurso más limitante o cuello de botella del sistema, que se denomina tambor. Para evitar que el tambor se quede sin trabajo o se sobrecargue, se establecen unos tiempos de protección o amortiguadores entre el tambor y los demás recursos, que se representan como una cuerda. El DBR permite optimizar el rendimiento del sistema, reducir los inventarios y los tiempos de entrega, y aumentar la flexibilidad y la satisfacción del cliente.

##*es decir:*##

tambor, amortiguador y cuerda, como se traduce el DBR.es una de las derivadas de la TOC donde explica que,
- el tambor es el ritmo con el que se mueve la produccion
- el amortiguador: es el recurso necesario para que las restricciones se mantengan operando, por ejemplo el inventario
- cuerda: es l sincornizacion necesaria para tirar de las unidades a través del sistema, como las señales *kanban*

**La regla SPT (Shortest Processing Time)** \\
es un algoritmo de planificación que asigna el tiempo de procesamiento a los trabajos en orden creciente de tiempo de proceso. **Ejemplo uno:** MG Auto tiene tres plantas: en Los Ángeles, Detroit y New Orleans; y dos centros principales de distribución en Denver y en Miami. Las capacidades de las tres plantas durante el próximo trimestre serán 1000, 1500 y 2000 autos. Las demandas trimestrales en los dos centros de distri- bución son 2300 y 1400 autos. El kilometraje entre las fábricas y los centros de distribución

- La empresa transportista cobra 8 centavos por KM y por auto. El costo de transporte por auto, en las distintas rutas y redondeado hasta el $ más próximo

- TABLA 1

 Nueva\ Orleans\ ($) & 1,275  & 850  \\

- TABLA 2

\hline

  Los\ Angeles(1) & 80$ & 215$ \\
\hline
 Detroit\ (2) & 100$  & 108$ \\
\hline
 Nueva\ Orleans (3) & 102$  & 68$  \\
\hline
\end{array}

**Resolviendo:**

 **Variables de decisión:**

- $x_{ij}$: Cantidad de autos a transportar desde la planta $i$ hasta el centro de distribución $j$.

**Función objetivo:**

- Minimizar el costo total de transporte, dado por:

$$\text{Min } Z = 0.08 \sum\limits_{i=1}^{3}\sum\limits_{j=1}^{2}x_{ij}d_{ij}$$

- Donde $d_{ij}$ es la distancia en kilómetros entre la planta $i$ y el centro de distribución $j$, según la tabla 1.

**Restricciones:**

- Capacidad de la planta:
$$\sum\limits_{j=1}^{2}x_{ij} \leq c_i para\ i = 1, 2, 3$$

- Donde $c_i$ es la capacidad de la planta $i$.

- Demanda del centro de distribución:
$$\sum\limits_{i=1}^{3}x_{ij} \geq d_j$ para $j = 1, 2$$

- Donde $d_j$ es la demanda del centro de distribución $j$, según el enunciado.

- Restricciones de no negatividad:
$$x_{ij} \geq 0\ para\ i = 1, 2, 3\ y\ j = 1, 2$$

- *Además, las variables de decisión deben ser enteras, ya que no se pueden transportar fracciones de autos.*





Resolviendo con Python y la libreria PuLp:

````python
# Importar la biblioteca PuLP
import pulp

# Crear un problema de minimización
problem = pulp.LpProblem("Transporte", pulp.LpMinimize)

# Crear las variables de decisión
x11 = pulp.LpVariable("Los Angeles a Denver", lowBound=0, cat='Integer')
x12 = pulp.LpVariable("Los Angeles a Miami", lowBound=0, cat='Integer')
x21 = pulp.LpVariable("Detroit a Denver", lowBound=0, cat='Integer')
x22 = pulp.LpVariable("Detroit a Miami", lowBound=0, cat='Integer')
x31 = pulp.LpVariable("New Orleans a Denver", lowBound=0, cat='Integer')
x32 = pulp.LpVariable("New Orleans a Miami", lowBound=0, cat='Integer')

# Definir la función objetivo
problem += 0.08 * (1000*x11 + 2690*x12 + 1250*x21 + 1350*x22 + 1275*x31 + 850*x32)

# Definir las restricciones de oferta y demanda
problem += x11 + x12 <= 2300, "Oferta Los Angeles"
problem += x21 + x22 <= 1500, "Oferta Detroit"
problem += x31 + x32 <= 2000, "Oferta New Orleans"
problem += x11 + x21 + x31 >= 2300, "Demanda Denver"
problem += x12 + x22 + x32 >= 1400, "Demanda Miami"

# Resolver el problema
problem.solve()

# Imprimir el resultado
print("Estado:", pulp.LpStatus[problem.status])
print("Costo total de transporte: $", round(pulp.value(problem.objective), 2))

for variable in problem.variables():
    print(variable.name, "=", variable.varValue)
````

##### Ejemplo 2:


En este modelo, tenemos tres orígenes ($O_1$, $O_2$ y $O_3$) y cuatro destinos ($D_1$, $D_2$, $D_3$ y $D_4$), y la tabla muestra los costos de transporte desde cada origen a cada destino. El objetivo es determinar la asignación óptima de los productos a los destinos, minimizando el costo total de transporte.


\begin{array}{c|cccc}
\\
\text{Origen} & D_1 & D_2 & D_3 & D_4 \\ \hline
O_1 & 4 & 5 & 2 & 7 \\
O_2 & 3 & 6 & 8 & 2 \\
O_3 & 9 & 4 & 1 & 3 \\
\end{array}


**Resolviendo el modelo:**

La función objetivo de este modelo se expresa como:

\begin{equation*}
\text{minimizar } Z = \sum_{i=1}^{3}\sum_{j=1}^{4}c_{ij}x_{ij}
\end{equation*}

Donde:

- $Z$: representa el costo total de transporte.
- $c_{ij}$: es el costo de transportar una unidad del origen $i$ al destino $j$.
- $x_{ij}$: es la cantidad de unidades transportadas del origen $i$ al destino $j$.

Las restricciones de capacidad se expresan como:

\begin{align*}
\sum_{j=1}^{4} x_{1j} &\leq 60 \
\sum_{j=1}^{4} x_{2j} &\leq 70 \
\sum_{j=1}^{4} x_{3j} &\leq 80
\end{align*}

**Donde:**

La restricción $\sum_{j=1}^{4} x_{ij} \leq C_i$ indica que la cantidad total de productos que se envían desde el origen $i$ no puede superar su capacidad máxima $C_i$.
Las restricciones de demanda se expresan como:

\begin{align*}
\sum_{i=1}^{3} x_{i1} &\geq 50 \
\sum_{i=1}^{3} x_{i2} &\geq 70 \
\sum_{i=1}^{3} x_{i3} &\geq 80 \
\sum_{i=1}^{3} x_{i4} &\geq 30
\end{align*}

**Donde:**

- La restricción $\sum_{i=1}^{3} x_{ij} \geq D_j$ indica que la cantidad total de productos que se envían al destino $j$ debe ser al menos igual a su demanda $D_j$


````python
import pulp

# Define el problema
problem = pulp.LpProblem("Transportation Problem", pulp.LpMinimize)

# Define las variables de decision
origins = ['O1', 'O2', 'O3']
destinations = ['D1', 'D2', 'D3', 'D4']
x = pulp.LpVariable.dicts('shipment', ((o, d) for o in origins for d in destinations), lowBound=0, cat='Integer')

# Define los objetos de las funciones
costs = {
    ('O1', 'D1'): 4,
    ('O1', 'D2'): 5,
    ('O1', 'D3'): 2,
    ('O1', 'D4'): 7,
    ('O2', 'D1'): 3,
    ('O2', 'D2'): 6,
    ('O2', 'D3'): 8,
    ('O2', 'D4'): 2,
    ('O3', 'D1'): 9,
    ('O3', 'D2'): 4,
    ('O3', 'D3'): 1,
    ('O3', 'D4'): 3
}
problem += pulp.lpSum([costs[o, d] * x[o, d] for o in origins for d in destinations])

# Define las restricciones
for o in origins:
    problem += pulp.lpSum([x[o, d] for d in destinations]) == 1
for d in destinations:
    problem += pulp.lpSum([x[o, d] for o in origins]) == 1

# resuelve el problema
problem.solve()

# imprime la solucion Optima
print("Optimal solution:")
for o in origins:
    for d in destinations:
        if x[o, d].value() > 0:
            print(f"Ship {x[o, d].value()} from {o} to {d}")
print(f"Total cost: {pulp.value(problem.objective)}")
````

#### Metodo grafico de localizacion


El método gráfico de localización de plantas es una técnica utilizada para seleccionar la ubicación óptima de una nueva planta o instalación. El proceso implica el uso de un mapa de la zona para identificar los factores críticos que influyen en la selección de la ubicación, como la disponibilidad de materias primas, la proximidad al mercado, la disponibilidad de mano de obra y la infraestructura existente.

1. Identificación de los factores críticos: El primer paso es identificar los factores críticos que influyen en la ubicación de la planta. Estos factores pueden variar según el tipo de planta y la industria. Por ejemplo, una planta de procesamiento de alimentos puede requerir una ubicación cerca de los campos de cultivo, mientras que una planta de fabricación de alta tecnología puede requerir una ubicación cerca de los centros de investigación y desarrollo.

2. Creación del mapa: Una vez que se han identificado los factores críticos, se crea un mapa de la zona que incluye información detallada sobre la ubicación de estos factores. El mapa también puede incluir información sobre la topografía, las vías de acceso, la disponibilidad de energía y agua, y otros aspectos relevantes.

3. Asignación de pesos a los factores críticos: Una vez que se ha creado el mapa, se asigna un peso a cada uno de los factores críticos en función de su importancia relativa. Esto permite que se le dé una mayor importancia a los factores más críticos en el proceso de toma de decisiones.

4. Análisis de las ubicaciones potenciales: Se identifican las posibles ubicaciones para la nueva planta o instalación y se evalúan en función de los factores críticos y sus respectivos pesos. Se pueden utilizar diferentes métodos para hacer esto, como análisis multicriterio o análisis de distancia.

5. Selección de la ubicación óptima: Finalmente, se selecciona la ubicación óptima en función de los resultados del análisis y se procede a la planificación y construcción de la planta.

#### sistema KANBAN

Uno de los aspectos de los sistemas esbeltos, y el TPS en particular, que ha recibido mayor divulgación es el sistema kanban desarrollado por Toyota.

- La expresión kanban, que en japonés significa “tarjeta” o “registro visible”, se refiere a las tarjetas que se utilizan para controlar el flujo de la producción en la fábrica. En el sistema kanban más elemental, se coloca una tarjeta en cada contenedor de artículos producidos. El contenedor almacena un porcentaje determinado de las necesidades diarias de producción de ese artículo. Cuando el usuario de esas partes vacía el contenedor, la tarjeta se retira del mismo y se coloca en un depósito de recepción. El contenedor vacío es llevado al área de almacenamiento, y la presencia de la tarjeta en el depósito indica que es necesario producir más de esas partes para llenar otro contenedor. Una vez que éste se vuelve a llenar, la tarjeta se coloca de nuevo en el contenedor, el cual se devuelve a un área de almacenamiento. El ciclo vuelve a empezar cuando el usuario de las partes recoge el contenedor con la tarjeta adjunta.

### la regla de Johnson

La regla de Johnson es un algoritmo heurístico utilizado para resolver situaciones de secuenciación de procesos que operan dos o más órdenes (operaciones) que pasan a través de dos máquinas o centros de trabajo. Su principal objetivo es minimizar el tiempo de procesamiento total del grupo de trabajos. Este algoritmo consiste en la aplicación de cuatro sencillos pasos:

- El primer paso consiste en listar todos los trabajos u operaciones junto con su tiempo de procesamiento por cada centro de trabajo o máquina.

- El segundo paso consiste en seleccionar el tiempo de procesamiento más corto. Recuerde que como resultado de la aplicación del paso 1 obtenemos dos columnas de tiempos (tiempos del centro de trabajo 1 y 2);
  - Sí el tiempo más corto se encuentra en la columna correspondiente al centro de trabajo (máquina) 1, este trabajo se programa primero, en caso contrario y de estar en la columna correspondiente al centro de trabajo 2, este trabajo se programa al final. Cualquier empate puede romperse de forma arbitraria.

- Una vez se programa un trabajo, sea al inicio o sea al final, este se elimina de la lista inicial.

- Se deben repetir los pasos 2 y 3 para los trabajos restantes, trabajando hacia el centro de la secuencia.

"La carga" libera trabajos a las instalaciones. Aunque la carga misma asigna trabajo a las instalaciones, no especifica el orden en que se deben realizar los trabajos en la instalación.

Los métodos de *secuenciación * determinan el orden de trabajo en la instalación.

#### Clasificación de problemas de programación
Los problemas de programación han sido clasificados de acuerdo a muchos criterios. Algunos de los criterios importantes se basan en
1. Secuencia de máquinas
2. Número de máquinas
3. Tiempos de procesamiento
4. Hora de llegada del trabajo
5. Funciones objetivas

#### Secuencia de Máquinas

Los problemas de programación se pueden clasificar como flow-shop y job-shop en función de la secuencia de máquinas requeridas por los trabajos a procesar.

Un flow-shop consta de varios trabajos y varias máquinas. Todos los trabajos requieren máquinas en el mismo orden para ser procesados.

un taller de flujo en el que se procesan tres trabajos A, B y C en cuatro máquinas M1, M2, M3 y M4.

Las secuencias de máquinas para procesar estos trabajos son las mismas (M1–M3–M4–M2).

En un job-shop, la secuencia de máquinas será mixta, es decir, los trabajos pueden requerir máquinas en diferentes secuencias.

las clasificaciones basicas son

- maquinas unicas
- maquinas dobles
- multiples maquinas



#### tiempo de procesamiento

Si los tiempos de procesamiento de todos los trabajos son conocidos y constantes, el problema de programación se denomina problema determinista. El problema de programación se llama probabilístico (o estocástico) si los tiempos de procesamiento no son fijos, es decir, los tiempos de procesamiento deben estar representados por una distribución de probabilidad.

#### tiempo de llegada de trabajos

Con base en este criterio, los problemas de programación se clasifican en problemas estáticos y dinámicos. En el caso de problemas estáticos, el número de trabajos es fijo y no cambiará hasta que se haya procesado el conjunto actual de trabajos. En el caso de problemas dinámicos, los nuevos trabajos ingresan al sistema y pasan a formar parte del conjunto actual de trabajos sin procesar. La tasa de llegada de trabajos se da en el caso de problemas dinámicos.

### Regla de Jhonson

La regla de Johnson es un método probado para dar una solución óptima.
- Hay cinco posiciones de secuencia 1–5. La regla de Johnson asigna cada trabajo a una de estas posiciones de manera óptima. Esta regla también requiere que se use la misma secuencia óptima en ambas máquinas.
- La regla también asume que no se permite preferencia ("no pasar"), lo que significa que una vez que se inicia un trabajo, no se puede interrumpir.
***
**Usamos el siguiente proceso de cuatro pasos para encontrar la secuencia óptima.**
***
1.  encuentre el tiempo mínimo de procesamiento considerando los tiempos en ambas máquinas.
2.  Identifique el trabajo correspondiente y la máquina correspondiente por el tiempo mínimo identificado en el Paso 1.
3.  regla de programación
  1.  Si la máquina identificada en el Paso 2 es la máquina M1, entonces la identificación del trabajo especificado en el paso 2 se programará en la primera posición de programación disponible.
  2.  Si la máquina identificada en el Paso 2 es la máquina M2, entonces el trabajo identificado en el Paso 2 se programará en la última posición de programación disponible.
  4.  Eliminar el trabajo de consideración cuya posición se ha fijado en Paso 3 y vaya al Paso 1.

## Programacion de una sola maquina

### Definicion:
Hay una sola máquina en la que se deben procesar varios trabajos. Es necesario especificar el orden en que se procesarán estos trabajos. Este cronograma no se cambiará hasta que se hayan procesado todos los trabajos.
 - Esta es la versión "estática" del problema.
 - En la versión “dinámica”, el horario se puede modificar.

#### Funcion objetivo:

Hay muchos criterios para evaluar los programas de producción.

- Consideramos minimizar el "make-span" mientras estudiamos los problemas de "flow-shop" de 2 máquinas.
  En el caso de un problema de una sola máquina, minimizar el "make-span" es irrelevante porque todas las secuencias darán el mismo "make-span", que es el tiempo de finalización del último trabajo de la secuencia.

Para un problema de una sola máquina, minimizar el tiempo de flujo promedio,
***
*cuál es la cantidad promedio de tiempo requerido para completar cada trabajo en el grupo*
***
este es un criterio más apropiado.

- Otra medida es el número promedio de puestos de trabajo en el sistema. Ambas medidas son funciones de la hora de inicio y la hora de finalización de cada trabajo.

presentamos otro término **"fecha de vencimiento"** (DUE-DATE) O (DEAD LINE) para cada trabajo. La fecha de *vencimiento de un trabajo* indica el tiempo en el que se debe completar un trabajo.

Tniendo en cuenta las reglas de programacion:

+ Recordamos:

1. **Por orden de llegada (FCFS)**. Los trabajos se procesan en el orden en que llegaron a la máquina.
2. **Tiempo de procesamiento más corto (SPT)**. Esto también se denomina tiempo de operación más corto. Entre los trabajos disponibles, el trabajo que requiere el tiempo de procesamiento mínimo se procesa primero y luego los otros trabajos se procesan en orden ascendente de sus tiempos de procesamiento.
3. **Fecha de vencimiento más temprana (más corta) (EDD)**. El trabajo que tiene la fecha de vencimiento más pequeña se procesa primero y luego los otros trabajos se procesan en orden ascendente de sus fechas de vencimiento.

# simulacion de montecarlo:

La simulación de Monte Carlo es una técnica computacional que utiliza muestreo aleatorio repetido para obtener resultados numéricos. Se emplea en una amplia variedad de problemas, desde finanzas hasta física, para modelar sistemas complejos y obtener estimaciones de cantidades que serían difíciles o imposibles de calcular de forma analítica.

## simulacion de montecarlo en python

en python podemos usar para simular librerias como SIMPY
la cual valida los entornos operativos del sistema, pero cuando a montecarlo nos referimos debemos optar por los aleatroios y la desripcion del sistema en funcion de los eventos.

por ejemplo
el lanzamiento de una moneda:

se usa la libreria "random" para la generacion de los aleatorios
y se usa la libreria " matplotlin" como ejeciutante de los calculos basicos del sistema

se usa la funcion "def" para definir la funcion a usar ene ste caso "simular_lanzamientos"

````python
import random
import matplotlib.pyplot as plt

def simular_lanzamientos(n_lanzamientos):
  resultados = []
  for _ in range(n_lanzamientos):
    resultado = random.choice(["Cara", "Cruz"])
    resultados.append(resultado)
  return resultados

# Simulación de 1000 lanzamientos
resultados = simular_lanzamientos(1000)

# Conteo de caras y cruces
caras = resultados.count("Cara")
cruces = resultados.count("Cruz")

# Visualización
plt.bar(["Cara", "Cruz"], [caras, cruces])
plt.title("Simulación de 1000 Lanzamientos de una Moneda")
plt.xlabel("Resultado")
plt.ylabel("Frecuencia")
plt.show()
````

Como hemos visto anteriormente, SJF es una estrategia sencilla y efectiva para minimizar el tiempo de finalización total. Consiste en ejecutar primero el trabajo con menor tiempo de procesamiento.

Ordenar los trabajos: Ordenamos los trabajos por tiempo de procesamiento de menor a mayor: C, A, D, B, E.
Asignar a la máquina: Asignamos los trabajos a la máquina siguiendo el orden obtenido: C, A, D, B, E.

````python
def sjf(jobs):
  jobs.sort(key=lambda x: x[1])  # Ordenar por tiempo de procesamiento
  start_time = 0
  for job, duration in jobs:
    print(f"Trabajo {job}: Inicio={start_time}, Fin={start_time + duration}")
    start_time += duration

# Datos de entrada
jobs = [('A', 3), ('B', 5), ('C', 2), ('D', 4), ('E', 6)]

# Ejecutar el algoritmo
sjf(jobs)
````

# Modelos deterministicos
****

In all the scheduling problems considered the number of jobs and the number of machines are assumed to be finite. The number of jobs is denoted by $n$ and the number of machines by $m$. Usually, the subscript $j$ refers to a job while the subscript $i$ refers to a machine. If a job requires a number of processing steps or operations, then the pair $(i,j)$ refers to the processing step or operation of job $j$ on machine $i$. The following pieces of data are associated with job $j$.


- Tiempo de procesamiento (pij)
El pij representa el tiempo de procesamiento del trabajo j en la máquina i. El subíndice i se omite si el tiempo de procesamiento del trabajo j no depende de la máquina o si el trabajo j solo debe procesarse en una máquina específica.

- Fecha de liberación (rj)
La fecha de liberación rj del trabajo j también puede denominarse como la fecha de disponibilidad. Es el momento en que el trabajo llega al sistema, es decir, el momento más temprano en el que el trabajo j puede comenzar su procesamiento.

- Fecha de entrega (dj)
La fecha de entrega dj del trabajo j representa la fecha de envío o finalización comprometida (es decir, la fecha prometida al cliente). Se permite la finalización de un trabajo después de su fecha de entrega, pero se incurrirá en una penalización. Cuando se requiere que se cumpla la fecha de entrega, se denomina una fecha límite y se representa con $d_j$

- Peso (wj)
El peso wj del trabajo j es básicamente un factor de prioridad, que denota la importancia del trabajo j en relación con los demás trabajos en el sistema. Por ejemplo, este peso puede representar el costo real de mantener el trabajo en el sistema. Este costo podría ser de almacenamiento o inventario; también podría representar el valor ya agregado al trabajo.

## Definiciones:

- **Definición:** (Programación sin demora):
Una programación factible se denomina sin demora si ninguna máquina permanece inactiva mientras una operación está esperando ser procesada.

Requerir que una programación sea sin demora equivale a prohibir la inactividad no forzada. Para muchos modelos, incluidos aquellos que permiten interrupciones y tienen funciones objetivo regulares, existen programaciones óptimas que son sin demora.

- **Definición:** (Programación Activa):
Una programación factible no preemptiva se denomina activa si no es posible construir otra programación, mediante cambios en el orden de procesamiento en las máquinas, en la que al menos una operación termine antes y ninguna operación termine después.

- **Definición** (Programación Semi-Activa).
Una programación factible no preemptiva se denomina semi-activa si ninguna operación puede completarse antes sin cambiar el orden de procesamiento en ninguna de las máquinas.

## PERT

El acrónimo PERT significa Program Evaluation and Review Technique, que traducido es Técnica de revisión y evaluación de programas. Proporciona una representación visual del cronograma de un proyecto y desglosa las tareas individuales

Starting about 1957, two similar approaches to large-scale project network plan- ning and tracking were begun at separate locations and for different reasons. These were:

1. PERT—program evaluation review technique
2. CPM—critical path method

**PERT** was developed by the US Navy Special Projects Office in conjunction with Booz Allen Hamilton for the Polaris submarine launched missile project. This cold war project was considered urgent by the government and time was a critical variable. There were about 100,000 activities divided amongst thousands of suppliers.

PERT set up activity networks, ideal for large projects, which could be systematically analyzed by computers.

**CPM** was a similar method developed by DuPont and Remington Rand, which later became Unisys. It was used to design and coordinate chemical plant opera- tions. Even at the time of development, computers were crucial.

## modelo VRP:



###**1. Generalidades del VRP**

*1.1 Historia*

El VRP tiene sus raíces en la década de 1950, cuando las empresas comenzaron a enfrentarse al desafío de optimizar la distribución de sus vehículos para atender a clientes y realizar entregas de manera eficiente. Aunque en ese entonces no se le denominaba VRP, este problema inicial sentó las bases para lo que se convertiría en un campo de estudio y aplicación crucial en la gestión logística.

En la década de 1960,  a través de prácticas, aplicaciones y estudios desarrollados investigadores como Dantzig y Ramser  representaron una aplicación real relacionada con la entrega de gasolina a las estaciones de servicio y propusieron la formulación matemática en el que formalizaron el VRP como un problema matemático, el cual surge como una generalización del problema clásico el agente viajero (TSP) en el cual un vendedor tiene que recorrer una serie de clientes una sola vez, para luego volver al lugar de partida.

Con el tiempo se han desarrollado diferentes variantes del VRP para resolver situaciones logisticas, las cuales van adaptadas demandas cambiantes del mercado.

*1.2 Definición*

El modelo clásico del  modelo de transporte conocido como VRP, consiste en determinar el diseño de rutas para una flota de vehiculos que parten de un deposito, almacen o bodega central y al cual tienen que regresar, luego de atender un conjunto de clientes dispersos en una zona geográfica, teniendo en cuenta las restricciones de capacidad de carga de los vehículos, distancia máxima recorrida por éstos, y respetando el horario de trabajo: todo ello con el fin de buscar el costo mínimo.

*1.3 Principales variantes del VRP*

En el caso de VRP, existen diferentes variantes que pueden considerar las capacidades de carga de los vehículos, ventanas de tiempo para la entrega de mercancías, múltiples depósitos de origen, entre otros aspectos que modifican la naturaleza del problema original. Estas variantes son importantes porque permiten adaptar el modelo de optimización a situaciones reales más complejas y específicas, mejorando así la capacidad de resolver problemas del mundo real de manera más efectiva, algunas de estas variantes son:

- **Problema de ruteo vehicular con capacidad (CVRP):**
El problema tradicional CVRP (por sus siglas en inglés
Capacitated Vehicle Routing Problem ó Problema de Ruteo de
Vehículos con Capacidad Limitada) puede ser descrito en su
forma más sencilla como una flota de vehículos con
capacidades uniformes que tiene que satisfacer la demanda de
un grupo de clientes a través de un conjunto de rutas que
empiezan y terminan en un almacén común y que representan el
menor costo posible, así como la identificación del orden de
visita a los mismos

-  **VRP con ventanas de tiempo:** Es un problema de VRP en el cual existe una restricción adicional de una ventana de tiempo establecida, la cual es el periodo entre las horas de inicio y fin en el que una ruta visitará una ubicación de red, es decir, una ventana describe cuándo un vehiculo puede llegar a la ubicación.En este sentido, el servicio de cada consumidor debe empezar dentro de la ventana de tiempo, en caso que el vehiculo llegue antes del tiempo establecido, este debe esperar hasta el instante de tiempo en que el serivicio deba empezar.  

- **Problema de ruteo vehicular con devoluciones:** En el
VRP con retornos (Backhauls) los consumidores pueden demandar o retornar algunas mercancías.

 En el problema tipo VRPB el conjunto de clientes se divide en dos. El primero contiene n centros de consumo con línea de recorrido (linehauls), que requieren que una cantidad dada de productos sea entregada y el segundo contiene m centros de consumo con recorrido de vuelta o retornos (backhauls), donde se requiere que una cantidad de productos dada deba ser recogida.

  En esta variación de VRP hay una restricción importante entre linehaul y backhaul, cuando una ruta pueda  servir a ambos tipos de consumidores todos los centros de consumo con linehauls deben ser atendidos antes  que algún centro de consumo con backhauls, si los hay.

####**Ejercicio variante CVRP**

La empresa de logística "Envíos Rápidos" tiene tres bodegas principales (A, B y C) desde donde distribuye productos a sus clientes. La cual cuenta con una flota de dos camiones repartidores con capacidad de carga de 30 unidades cada uno. Las distancias de las bodegas entre las bogedas se muestran en la siguiente matriz:

\begin{array}{|c|c|c|c|}
\hline
\textbf{ Bodegas} & \textbf{A} & \textbf{B} & \textbf{C}\\
\hline
  A & 0 & 5 & 10\\
\hline
 B  & 5 & 0 & 8\\
\hline
 C & 10 & 8 & 0\\
\hline
\end{array}

\\

La demanda de los clientes en cada bodega es:
- Bodega A: 10 unidades
- Bodega B: 15 unidades
- Bodega C: 20 unidades

En este sentido, se solicita determinar las rutas óptimas para los dos camiones que minimicen la distancia total recorrida, satisfaciendo la demanda de cada bodega sin exceder la capacidad de los vehículos.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
import pulp

# Solicitar la cantidad de camiones al usuario
cantidad_camiones = int(input("Ingrese la cantidad de camiones: "))

# Solicitar la capacidad en kg de los camiones al usuario
capacidad_camiones = float(input("Ingrese la capacidad en kg de los camiones: "))

# Solicitar la cantidad de clientes al usuario
cantidad_clientes = int(input("Ingrese la cantidad de clientes: "))

# Inicializar listas para almacenar los nombres y las coordenadas de los clientes
clientes = []
coordenadas = []

# Agregar el depósito "D" en el origen del plano cartesiano
clientes.append("D")
coordenadas.append((0, 0))

# Generar coordenadas aleatorias para cada cliente
np.random.seed(42)  # Para reproducibilidad
for i in range(1, cantidad_clientes + 1):
    x = np.random.uniform(1, 100)
    y = np.random.uniform(1, 100)
    clientes.append(f"cliente_{i}")
    coordenadas.append((x, y))

# Convertir la lista de coordenadas a un array numpy
coordenadas = np.array(coordenadas)

# Calcular la matriz de distancias
matriz_distancias = cdist(coordenadas, coordenadas, metric='euclidean')

# Mostrar las ubicaciones de los puntos en un plano cartesiano
plt.figure(figsize=(10, 6))
for i, (x, y) in enumerate(coordenadas):
    plt.scatter(x, y, label=clientes[i])
    plt.text(x, y, clientes[i], fontsize=12, ha='right')
plt.title('Ubicaciones de los Clientes y el Depósito')
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.grid(True)
plt.legend()
plt.show()

# Mostrar la matriz de distancias
print("Matriz de Distancias:")
print(matriz_distancias)

# Solicitar la demanda de cada cliente
demandas = [0]  # La demanda del depósito es 0
for i in range(1, cantidad_clientes + 1):
    demanda = float(input(f"Ingrese la demanda del {clientes[i]} en kg: "))
    demandas.append(demanda)

# Crear el problema de optimización
prob = pulp.LpProblem("CVRP", pulp.LpMinimize)

# Variables de decisión: x[i][j][k] será 1 si el camión k viaja de i a j, 0 en caso contrario
x = [[[pulp.LpVariable(f"x_{i}_{j}_{k}", cat='Binary') for k in range(cantidad_camiones)] for j in range(cantidad_clientes + 1)] for i in range(cantidad_clientes + 1)]

# Variables auxiliares: u[i] representa el orden de visita del cliente i para evitar subtours
u = [pulp.LpVariable(f"u_{i}", lowBound=0, cat='Continuous') for i in range(cantidad_clientes + 1)]

# Función objetivo: minimizar la distancia total recorrida
prob += pulp.lpSum(matriz_distancias[i][j] * x[i][j][k] for k in range(cantidad_camiones) for i in range(cantidad_clientes + 1) for j in range(cantidad_clientes + 1))

# Restricciones
# Cada cliente debe ser visitado exactamente una vez por exactamente un vehículo
for j in range(1, cantidad_clientes + 1):
    prob += pulp.lpSum(x[i][j][k] for i in range(cantidad_clientes + 1) for k in range(cantidad_camiones)) == 1

# Cada vehículo debe salir del depósito
for k in range(cantidad_camiones):
    prob += pulp.lpSum(x[0][j][k] for j in range(1, cantidad_clientes + 1)) == 1

# Cada vehículo que llega a un cliente debe salir del mismo
for k in range(cantidad_camiones):
    for j in range(1, cantidad_clientes + 1):
        prob += pulp.lpSum(x[i][j][k] for i in range(cantidad_clientes + 1)) == pulp.lpSum(x[j][i][k] for i in range(cantidad_clientes + 1))

# Capacidad de los vehículos no debe ser excedida
for k in range(cantidad_camiones):
    prob += pulp.lpSum(demandas[j] * x[i][j][k] for i in range(cantidad_clientes + 1) for j in range(1, cantidad_clientes + 1)) <= capacidad_camiones

# Subtour elimination
for i in range(1, cantidad_clientes + 1):
    for j in range(1, cantidad_clientes + 1):
        if i != j:
            for k in range(cantidad_camiones):
                prob += u[i] - u[j] + (cantidad_clientes + 1) * x[i][j][k] <= cantidad_clientes

# Resolver el problema
prob.solve()

# Resultados
print("Estado de la solución:", pulp.LpStatus[prob.status])

for k in range(cantidad_camiones):
    ruta = []
    distancia_recorrida = 0
    for i in range(cantidad_clientes + 1):
        for j in range(cantidad_clientes + 1):
            if pulp.value(x[i][j][k]) == 1:
                ruta.append((clientes[i], clientes[j]))
                distancia_recorrida += matriz_distancias[i][j]
    print(f"\nCamión {k + 1}:")
    print("Ruta:", ruta)
    print(f"Distancia recorrida: {distancia_recorrida:.2f} km")

# Resumen de distancias recorridas
print("\nResumen de distancias recorridas por cada camión:")
for k in range(cantidad_camiones):
    distancia_recorrida = sum(matriz_distancias[i][j] * (pulp.value(x[i][j][k]) if pulp.value(x[i][j][k]) is not None else 0) for i in range(cantidad_clientes + 1) for j in range(cantidad_clientes + 1))
    print(f"Camión {k + 1}: {distancia_recorrida:.2f} km")
````

### TOC con simpy

5 maquinas en paralelo
dos productos p y q
tiempos de procesamiento de p: aleatorio entre 10 y 15 min
tiempo de procesamiento de q: aleatorio entre 9 y 16



````python
import simpy
import random
import matplotlib.pyplot as plt

def process_part(env, machine, part, processing_time):
    with machine.request() as req:
        yield req
        yield env.timeout(processing_time)

def generate_parts(env, machine, part_type, num_parts, processing_time_range):
    for i in range(num_parts):
        yield env.timeout(random.expovariate(1/5))  # Tiempo entre llegadas
        env.process(process_part(env, machine, part_type, random.uniform(*processing_time_range)))

def simulate(env, machines, part_types): #Removed num_parts and processing_time_ranges as arguments
    for i, machine in enumerate(machines):
        for part_type, num_parts, processing_time_range in part_types: #Access num_parts and processing_time_range directly from part_types
            env.process(generate_parts(env, machine, part_type, num_parts, processing_time_range))
    env.run()

# Parámetros
num_machines = 5
machines = [simpy.Resource(env, capacity=1) for _ in range(num_machines)]
part_types = [
    ('P', 10, (10, 15)),  # Parte P, 10 unidades, tiempo entre 10 y 15 minutos
    ('Q', 10, (9, 16))   # Parte Q, 10 unidades, tiempo entre 9 y 16 minutos
]

# Crear el entorno de simulación
random.seed(42)
env = simpy.Environment()
simulate(env, machines, part_types) #Removed num_parts and processing_time_ranges from the function call
````

###TOC con Pyomo

````python
from pyomo.environ import *

# Crear un modelo
model = ConcreteModel()

# Definir las variables de decisión
model.x = Var(within=NonNegativeReals)  # Cantidad de producto A
model.y = Var(within=NonNegativeReals)  # Cantidad de producto B

# Definir la función objetivo (maximizar ganancias)
model.obj = Objective(expr=10*model.x + 15*model.y, sense=maximize)

# Definir las restricciones
model.constraint1 = Constraint(expr=2*model.x + 3*model.y <= 100)  # Restricción de materia prima
model.constraint2 = Constraint(expr=model.x + model.y <= 80)  # Restricción de tiempo de producción

# Resolver el modelo
solver = SolverFactory('glpk')
results = solver.solve(model)

# Mostrar los resultados
print(results)
````
